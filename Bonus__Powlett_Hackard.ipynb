```python
# Import the dependencies.
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine
from log_inf import password
```

```python
# Create Database Connection

engine = sqlalchemy.create_engine(sqlalchemy.engine.url.URL.create(
        drivername="postgresql",
        username="postgres",
        password= password,
        host="localhost",
        port="5432",
        database="Pewlett_Hackard_db",
    ),
    echo_pool=True,
)

conn = engine.connect()
```

```python
#List the employee number, last name, first name, sex, and salary of each employee.

question1="""SELECT e.emp_no, e.last_name, e.first_name, e.sex, s.salary
FROM employees e
INNER JOIN salaries s ON e.emp_no = s.emp_no
ORDER BY e.emp_no;"""

df_question1=pd.read_sql_query(question1, conn)
print(len(df_question1))
df_question1.head()
```

```python
#List the first name, last name, and hire date for the employees who were hired in 1986.
question2= """SELECT first_name, last_name, hire_date
FROM employees
WHERE EXTRACT(YEAR FROM hire_date) = 1986
ORDER BY hire_date;"""

df_question2=pd.read_sql_query(question2, conn)
print(len(df_question2))
df_question2.head()
```

```python
# List the manager of each department along with their department number, department name, employee number, last name, and first name.

question3 = """SELECT dm.dept_no, d.dept_name, e.emp_no, e.last_name, e.first_name
from dept_manager dm
INNER JOIN departments d on dm.dept_no = d.dept_no
INNER JOIN employees e on dm.emp_no = e.emp_no;"""

df_question3 = pd.read_sql_query(question3, conn)
print(len(df_question3))
df_question3.head()
```

```python
#List the department number for each employee along with that employee’s employee number, last name, first name, and department name.

question4 = """SELECT d.dept_no, e.emp_no, e.last_name, e.first_name, d.dept_name
FROM employees e
INNER JOIN dept_emp de on de.emp_no = e.emp_no
INNER JOIN departments d on d.dept_no = de.dept_no;"""

df_question4 = pd.read_sql_query(question4, conn)
print(len(df_question4))
df_question4.head()
```

```python
#List first name, last name, and sex of each employee whose first name is Hercules and whose last name begins with the letter B.

question5 ="""SELECT first_name, last_name, sex
FROM employees
WHERE first_name = 'Hercules' and  last_name ~ '^B'
ORDER BY last_name;"""

df_question5 =pd.read_sql_query(question5, conn)
print(len(df_question5))
df_question5.head()
```

```python
#List each employee in the Sales department, including their employee number, last name, and first name.

question6= """SELECT d.dept_name, e.emp_no, e.last_name, e.first_name 
FROM employees e
INNER JOIN dept_emp de on de.emp_no = e.emp_no
INNER JOIN departments d on d.dept_no = de.dept_no
WHERE dept_name = 'Sales'
ORDER BY e.emp_no;"""

df_question6 =pd.read_sql_query(question6, conn)
print(len(df_question6))
df_question6.head()
```

```python
#List each employee in the Sales and Development departments, including their employee number, last name, first name, and department name.

question7="""SELECT d.dept_name, e.emp_no, e.last_name, e.first_name 
FROM employees e
INNER JOIN dept_emp de on de.emp_no = e.emp_no
INNER JOIN departments d on d.dept_no = de.dept_no
WHERE dept_name in ('Sales', 'Development')
ORDER BY e.emp_no; """

df_question7 =pd.read_sql_query(question7, conn)
print(len(df_question7))
df_question7.head()
```

```python
#List the frequency counts, in descending order, of all the employee last names (that is, how many employees share each last name).

question8="""SELECT last_name, COUNT(last_name) AS Frequency
FROM employees
GROUP BY last_name
ORDER BY Frequency DESC;"""

df_question8 =pd.read_sql_query(question8, conn)
print(len(df_question8))
df_question8.head()
```
